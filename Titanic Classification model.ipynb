```python
sns.boxplot(data=dataset, y='Age', x='Pclass')
# Middle Line represents the Mean value
```

```python
# To replace NULL values with Mean values of corresonding "Pclass"
def replace(cols):
    age = cols[0]
    Pclass = cols[1]
    if pd.isnull(age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 30
        elif Pclass==3:
            return 25
        else:
            return 30
    else:
        return age
```

```python
dataset[['Age','Pclass']].apply(replace)
# This is similar to put the dataset into the function
```

```python
dataset['Age'] = dataset[['Age','Pclass']].apply(replace,axis=1)
# The values are passed as a tuple, that's why we have used only one variable in function arguments
```

# Classification

```python
y=dataset['Survived']
```

```python
dataset.columns
```

```python
X = dataset[['Pclass','Age','Sex','SibSp','Parch','Embarked']]
```

```python
# string => number(label encoding) : Dummy variable => one hot encoding , remove one column
```

```python
sex = dataset['Sex']
sex = pd.get_dummies(sex, drop_first=True)
# We do one-hot encoding to remove multi-collinearity
```

```python
pclass = dataset['Pclass']
pclass = pd.get_dummies(pclass, drop_first=True)
```

```python
sibsp = dataset['SibSp']
sibsp = pd.get_dummies(sibsp, drop_first=True)
```

```python
parch = dataset['Parch']
parch = pd.get_dummies(parch, drop_first=True)
```

```python
embarked = dataset['Embarked']
embarked = pd.get_dummies(embarked, drop_first=True)
```

```python
X
# we don't need all these data
```

```python
# parch, embarked, etc. all are pandas type datatype
# To convert them , we will use 'concat()' method
pd.concat([age, embarked, parch, sibsp, pclass, sex])

# Not done properly, because its done Row-wise by default
```

```python
X = pd.concat([age, embarked, parch, sibsp, pclass, sex], axis=1)
# make sure that 'age' column do not have 'NaN' values
```

```python
X
```

```python
# Apply logictic regression ( for binary classification)
from sklearn.linear_model import LogisticRegression
```
